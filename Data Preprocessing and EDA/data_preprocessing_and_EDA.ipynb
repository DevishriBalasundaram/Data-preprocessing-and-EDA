### Univariate Analysis

Explore variables one by one and analyse by its type(categorical or continuous)

### Bivariate Analysis
Finds out relationship between two variables of any combination.                                                                
possible combinations: (continuous&continuous)(continuous&categorical)(categorical&categorical)  

To find relationship between continuous&continuous

- Scatter plot gives relationship between the variables
- To find strength we use corelation     

To find relationship between categorical&categorical
- Chi Sqr test derives the statistical significance of relationship between the variables   

To find relationship between continuous&categorical
- We draw box plots and z-test or t-test or ANOVA gives the statistical significance

### Missing value and its treatment
Missing value in a dataset leads to wrong prediction. So methods to treat are:
- Deletion (The threshold value for deleting the outliers is 2% of the entire data)
- Fill missing values:                                                                                                         
For numerical var: using mean or median                                                                                         
For categorical var: using mode
- Prediction model
- KNN Imputation

### Outlier detection and its treatment
Outlier is an observation that appears far away and diverges from an overall pattern in a sample

Univariate outlier: Found at distribution of a single variable (mostly use box plot)                                           
Bivariate outlier: Outliers in n-dimensional space.ie) distribution in multi dimensions (scatter plot)
    
Outliers are caused by natural and non natural errors

OUTLIER DETECTION:                                                                                                             
    Any value beyond the range of -1.5 * IQR to +1.5 * IQR                                                                     
    Any value out of range of 5th and 95th percentile                                                                           
    3 or more STD dev away from mean 
    
OUTLIER TREATMENT:                                                                                                             
   Deleting observations - if it is due to natural or small data entry errors.(The threshold value for deleting the outliers is 2% of the entire data)                                                   
   Transformation - which makes our distribution to be close to normal and removes outliers(log) and binning(variable transformation)                                                                                                                 
   Imputation - With mean,mode and median

### Feature Engineering
It is an art of extracting or attaining more information from the existing dataset which we have.(bringing out information from data)

#### Feature Transformation
Replacement of a variable (x) by a function. ie) square,cube root,log                                                           
After transformation it changes the distribution ie) If the data is highly skewed, after transformation it becomes normal distribution.

***When transformations are used:

- change the scale (Standardize the values of the variable,for better understanding)
- Transform non linear into linear relationships (for easier comparision - commonly log transformation is used)
- Remove skewness inorder to have normal distribution (for right skewed distribution - sq/cube/log) (for left skew - sq/cube/exponential)  

***Other common approach used in implementation:

- Binning of variables ( When we have corelation,binning is used to categorize variables eg:- <20, 20-30, >30

#### Feature Scaling
It is a technique to standardize the independent features present in the data in a fixed range.

` Techniques to perform Feature Scaling:`
- Min-Max Normalization (values are shifted and rescaled so that they end up ranging between 0 and 1)
- Standardization (where the values are centered around the mean with a unit standard deviation)

Normalization is good to use when you know that the distribution of your data does not follow a Gaussian distribution. This can be useful in algorithms that do not assume any distribution of the data like K-Nearest Neighbors and Neural Networks.

Standardization, on the other hand, can be helpful in cases where the data follows a Gaussian distribution. However, this does not have to be necessarily true. Also, unlike normalization, standardization does not have a bounding range. So, even if you have outliers in your data, they will not be affected by standardization.

### Handling class imbalance
Class imbalance problems occurs mostly in classification problem where the classes are not represented equally.                                                               
eg:- For patients who do not have the disease, you’d have 100% accuracy.                                                       
     For patients who do have the disease, you’d have 0% accuracy.                                                             
After performing evaluation metrics (precision,recall,acc,f1 score) our one of our class may gives very less accuracy hence we need to handle the class imbalance

`Ways to handle class imbalance`
- Up-sample Minority Class (class having less data points will increase)
- Down-sample Majority Class (Reduces the no of data points which has more rows)
- Change Performance Metric (Area Under ROC Curve,Kappa)
- Penalize Algorithms (Cost-Sensitive Training) (SVM algorithm)
- Try Resampling the Dataset
- Use Tree-Based Algorithms: Their hierarchical structure allows them to learn signals from both classes.(DT,RF,Boosting)
